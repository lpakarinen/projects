**Decision tree** is a white box machine learning algorithm in which the individual targets (in this case, people) are passed through different nodes (input variables) of the tree. In each node the target is classified according to its attributes and the route splits into two depending on the “decision”. The goal is to divide the targets into as different categories (leaves) as possible. The algorithms can be used to both classify different targets (in this case, customers) and predict continuous values. In other words, a decision tree can be used for both classification and regression. Also, even though in this case the problem is binary (yes or no), decision tree is able to handle multi-output problems unlike for example logistic regression. The visualisation of a decision tree is similar to a tree drawn upside down, with its roots on the top and leaves at the bottom, hence its name. (Ray, 2017)

Unfortunately decision trees are infamously susceptible to overfitting and clinging too much to their training data. Therefore, after successfully creating a tree, the next step is usually to prune it to make the model simpler, prettier and more efficient. Pruning the tree drastically lowers the chance for errors using testing data, which also makes the whole algorithm more adaptable for future use. Pruning can be done either while or after creating the tree. Pruning the tree during its growth is called early stopping or pre-pruning, and there are multiple methods available for that. In one of them for example the tree checks the cross-validation error and stops its growing if the error is not high enough. The method max_leaf_nodes limits the number of leaves the decision tree can have and therefore controls its growth. Pruning the tree after it’s fully grown can be done in different methods, using for example cost complexity pruning, reduced error pruning or minimum error pruning. Pre-pruning and post-pruning methods can also be used simultaneously. (Yildirim, 2020)

As we can most likely later see, the model is not working without optimization as it clearly overfits the data. For optimization we use cost complexity pruning, because we found manually restricting the size of the tree time-consuming and inefficient. In CCP the idea is to find an ideal size for the tree by creating different versions, cross-validating them and picking the best and optimal one. In the code this is done by finding the ideal Alpha, which is a number that represents the amount of pruning the tree gets. To find the value for ideal alpha we need to use the smaller data set since the full dataset is too massive for the process using the equipment we have. (Starmer, 2020)
 
**Random forest** is a model that consists of several individual decision trees, and it gives the most popular prediction out of the different predictions made by the trees. The process is similar to cross-validation. Random forest algorithms can be used for solving both regression and classification problems, making it a versatile and powerful data-analytic tool. 

The trees are formed using m out of M input variables (M being the amount of the variables as a whole and m << M) and the trees do not correlate with each other. Unlike usually when using a single decision tree, in random forest the trees are not pruned at all, which leads to overfitting of the given training data. However, when all the trees use small, different parts of the dataset it averages itself out and reduces the variance of different outcomes. (Donges, 2019) 
Random forest can also offer information on which variables and decisions were the most important during the process. (Yiu, 2019) The algorithm measures a single variable's importance by sorting out how much the tree nodes that use a certain variable reduce impurity across all trees in the forest (Donges, 2019).

In this work we are specifically using random forest’s ability to find variable importance. It is used to discover which customer attributes are the most and least important when it comes to their decision on whether or not to subscribe for a term deposit. The data will be plotted to a barplot visualizing clearly which attributes matter and which do not. 
 
**Logistic regression** is a machine learning model which is useful for solving binary classification problems. The logistic regression model predicts the probability of an outcome. There are 3 types of logistic regression:

**1)** binary logistic regression such as spam or not spam email classification

**2)** multinomial logistic regression such as predicting the type of a wine

**3)** ordinal logistic regression such as a restaurant rating from 1 to 5 
(Navlani, 2019). 

The output of the linear regression is passed through an activation function called sigmoid function which maps real values between 0 and 1 (Agarwal, 2018). The value of the sigmoid function ranges between 0 and 1. The output value-y is 0.5 at X=0. 0.5 is used as the probability threshold to determine the classes (Agarwal, 2018) . If the probability is > 0.5, it is classified as as y=1 and if probability is >0.5, it is classified as y=0 (Agarwal, 2018). 
Before logistic regression model can be trained on the selected features, some preparations must be made. The dependent variable y must be categorical, while independent variables need to be independent from one another to avoid multicollinearity (Agarwal, 2018).

In this project’s case, we want to predict the probability of occurrence of the target variable y, which means whether or not the client subscribes to a term deposit. The target variable y is binary ("yes" =  1, or "no" =  0). Logistic regression model indicates which independent variables affect the outcome, the target variable y, and how strong is the connection between the feature variables and the target variable (Ray, 2017). We will implement classifier using scikit-learn’s Logistic Regression package.  Behind the scene, regularization is taken care of to prevent the model from overfitting the data (Agarwal, 2018).
 
**Support vector machine SVM** is one of the widely used supervised machine learning algorithms, which can be used in various classification use cases such as email classification, face detection, , intrusion detection (Navlani, 2019), and regression problems as well. 
SVM can provide higher accuracy than other classifiers such as logistic regression or decision trees (Navlani, 2019). SVM uses its kernel trick to handle nonlinear input spaces (Navlani, 2019). 

How SVM works is that the classifier divides data points using a hyperplane with the largest amount of margin (Navlani, 2019). SVM classifier is also referred to as a discriminative classifier and it looks for  an optimal hyperplane to classify new data points in an iterative way (Navlani, 2019).The key idea of SVM is to find a maximum marginal hyperplane (MMH) that best divides the dataset into classes (Navlani, 2019).

In SVM, support vectors are the data points, which are nearest to the hyperplane (Navlani, 2019), and a hyperplane is a decision plane which segregates a set of objects having different class memberships (Navlani, 2019).

Each data item is plotted as a point into an n-dimensional space (n = number of features), and where the data item’s value for each feature is its coordinate for the axel. Then SVM detects a hyperplane (whose dimension is n - 1) between the classes of data points, and it should be as far from the points as possible. (Gandhi, 2018). 

Advantages of SVM classifiers are that they provide good accuracy and fast prediction, and use less memory by taking a subset of training points in the decision phase. SVM works best with a clear margin of separation and in high dimensional space (Navlani, 2019).

Disadvantages of SVM are, however, that it works poorly with large datasets due to high training time, also with overlapping classes and it is sensitive to the kernel type used (Navlani, 2019).

```python
# Import PyDrive and associated libraries.
# This only needs to be done once per notebook.
from pydrive.auth import GoogleAuth
from pydrive.drive import GoogleDrive
from google.colab import auth
from oauth2client.client import GoogleCredentials

# Authenticate and create the PyDrive client.
# This only needs to be done once per notebook.
auth.authenticate_user()
gauth = GoogleAuth()
gauth.credentials = GoogleCredentials.get_application_default()
drive = GoogleDrive(gauth)

# Download the files based on its file ID.
# A file ID looks like: laggVyWshwcyP6kEI-y_W3P8D26sz
file_id = '1CqedNzdGmgpQdPY-E9M890TEZKtq3vxF'
downloaded = drive.CreateFile({'id': file_id})
downloaded.GetContentFile("bank_full.csv")

file_id_small = '1J55IYX9QL7i8YE1dvwjCc_-RjfPs-yMi'
downloaded_small = drive.CreateFile({'id': file_id_small})
downloaded_small.GetContentFile("bank.csv")


# Import all libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
#from sklearn.pipeline import make_pipeline
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import LabelEncoder
from sklearn.tree import DecisionTreeClassifier
from sklearn import metrics
from sklearn import tree
from sklearn.model_selection import GridSearchCV
from sklearn.externals.six import StringIO
import pydot
from sklearn import preprocessing
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import plot_confusion_matrix
import seaborn as sns
from sklearn.svm import SVC
from sklearn.pipeline import Pipeline
#from imblearn.pipeline import Pipeline 
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.compose import make_column_transformer
from sklearn.model_selection import cross_val_score

from imblearn.over_sampling import RandomOverSampler

#Read file including the datasets
df = pd.read_csv('bank_full.csv', delimiter=';' )
df_small = pd.read_csv('bank.csv', delimiter=';' )

#For later decoding, snipped from here: https://github.com/scikit-learn/scikit-learn/issues/12525

def get_column_names_from_ColumnTransformer(column_transformer):    
    col_name = []
    for transformer_in_columns in column_transformer.transformers_[:-1]:#the last transformer is ColumnTransformer's 'remainder'
        raw_col_name = transformer_in_columns[2]
        if isinstance(transformer_in_columns[1],Pipeline): 
            transformer = transformer_in_columns[1].steps[-1][1]
        else:
            transformer = transformer_in_columns[1]
        try:
            names = transformer.get_feature_names()
        except AttributeError: # if no 'get_feature_names' function, use raw column name
            names = raw_col_name
        if isinstance(names,np.ndarray): # eg.
            col_name += names.tolist()
        elif isinstance(names,list):
            col_name += names    
        elif isinstance(names,str):
            col_name.append(names)
    return col_name


def convert_datatypes(df):
    #change "yes" to 1 and "no" to 0
    df.default=df.default.map(dict(yes=1, no=0))
    df.housing=df.housing.map(dict(yes=1, no=0))
    df.loan=df.loan.map(dict(yes=1, no=0))
    df.y=df.y.map(dict(yes=1, no=0))

    #declare the right datatype for every column in a dictionary
    convert_dict = {'job': "category",
                    'marital': "category",
                    'education':"category",
                    'default':"bool",
                    'housing':"bool",
                    'loan':"bool",
                    'contact':"category",
                    'month':"category",
                    'poutcome':"category",
                    'y':"bool"
                  }
      
    #change the datatype
    df = df.astype(convert_dict)
      
    #check the datatype
    #df.dtypes

    #check the head
    df.head()
    return df

#Repeat the same for smaller dataset
df = convert_datatypes(df)
df_small = convert_datatypes(df_small)
```

###Analytics

```python
### Preparation ###

#Declare Attributes & Label
# Attributes
X = df[['job','marital','education','age','default','balance','housing', 'loan']]
X_small = df_small[['job','marital','education','age','default','balance','housing', 'loan']]

# Label
y = df['y']
y_small = df_small['y']

# Oversampling
ros = RandomOverSampler(random_state=0)
Xs,ys = ros.fit_resample(X, y)
Xs_small,ys_small = ros.fit_resample(X_small, y_small)

X = pd.DataFrame(Xs, columns = ['job','marital','education','age','default','balance','housing', 'loan'])
y = pd.Series(ys)

X_small = pd.DataFrame(Xs_small, columns = ['job','marital','education','age','default','balance','housing', 'loan'])
y_small = pd.Series(ys_small)

convert_dict2 = {'job': "category",
                    'marital': "category",
                    'education':"category",
                    'age':"int",
                    'default':"bool",
                    'balance':"int",
                    'housing':"bool",
                    'loan':"bool",
#                  'contact':"category",
#                    'month':"category",
#                   'poutcome':"category",
#                   'y':"bool"
                  }
X = X.astype(convert_dict2)
X_small = X_small.astype(convert_dict2)
```

```python
# Splitting into Training & Test Data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
X_train_small, X_test_small, y_train_small, y_test_small = train_test_split(X_small, y_small, test_size=0.2)

# Encode Categorical Data, Scale Numerical Data
features_to_scale = list(X_train.select_dtypes(include = ['int64']).columns)
features_to_encode = list(X_train.select_dtypes(include = ['category']).columns)


preprocessor = ColumnTransformer(
    transformers=[
       ('num', StandardScaler(), features_to_scale),
       ('cat', OneHotEncoder(), features_to_encode)
])

X_train_small_pre = pd.DataFrame(preprocessor.fit_transform(X_train_small).toarray())
X_test_small_pre = pd.DataFrame(preprocessor.fit_transform(X_test_small).toarray())
```

#### Random Forest

```python
### Random Forest ###

# https://towardsdatascience.com/my-random-forest-classifier-cheat-sheet-in-python-fedb84f8cf4f


# Classifier
rf_classifier = RandomForestClassifier(
                      min_samples_leaf=50,
                      n_estimators=150,
                      bootstrap=True,
                      oob_score=True,
                      n_jobs=-1,
                      max_features='auto')

# Pipeline
pipe_rf = Pipeline(steps=[('preprocessor', preprocessor),
                          ('classifier', rf_classifier)])

# fit the Pipeline
pipe_rf.fit(X_train, y_train)

# Predict Labels
y_pred_rf = pipe_rf.predict(X_test)

# Accuracy and Confusion matrix
print(f"Accuracy of the RF Model: {accuracy_score(y_test,y_pred_rf)} %")

plot_confusion_matrix(pipe_rf, X_test, y_test, display_labels= ["Not subscribed", "Subscribed"])  
plt.show()

#Accuracy of the RF Model: 0.6657899680631223 %
```

```python
### Feature Importance ###
# Which features are important? What distinguishes a willing and an unwilling customer?   
X_train_encoded = get_column_names_from_ColumnTransformer(preprocessor)

feature_importances = list(zip(X_train_encoded, rf_classifier.feature_importances_))

# Rankinglist(zip(X_train_encoded, rf_classifier.feature_importances_)
feature_importances_ranked = sorted(feature_importances, key = lambda x: x[1], reverse = True)

# Print out the feature and importances
[print('Feature: {:35} Importance: {}'.format(*pair)) for pair in feature_importances_ranked];

feature_names_15 = [i[0] for i in feature_importances_ranked[:15]]
y_ticks = np.arange(0, len(feature_names_15))
x_axis = [i[1] for i in feature_importances_ranked[:15]]

plt.figure(figsize = (10, 14))
plt.barh(feature_names_15, x_axis)   

#horizontal barplot
plt.title('Random Forest Feature Importance (Top 15)')
plt.xlabel('Features',fontdict= {'fontsize' : 16})

#plt.show()
```

#### Decision Tree


```python
#Create primary Decision Tree

dt_classifier = DecisionTreeClassifier()

pipe_dt = Pipeline(steps=[('preprocessor', preprocessor),
                          ('classifier', dt_classifier)])

# fit the Pipeline
pipe_dt.fit(X_train, y_train)

# Predict Labels
y_pred_dt = pipe_dt.predict(X_test)

# Accuracy
print(f"Accuracy of the DT Model: {accuracy_score(y_test,y_pred_dt)}")

plot_confusion_matrix(pipe_dt, X_test, y_test, display_labels= ["Not subscribed", "Subscribed"])  
plt.show()
```

```python
#Visualize Preliminary Decision Tree
fn = get_column_names_from_ColumnTransformer(preprocessor)
cn = ['Not_subscribed', 'Subscribed']

fig, axes = plt.subplots(nrows = 1,ncols = 1,figsize = (19,6), dpi=1000)
tree.plot_tree(dt_classifier, 
                feature_names=fn,  
                class_names=cn,
                filled=True)
```

```python
# https://github.com/iasarthak/My-Data-Logs/blob/master/Concepts/Cost%20Complexity%20Pruning.ipynb

path=dt_classifier.cost_complexity_pruning_path(X_train_small_pre, y_train_small)
alphas=path['ccp_alphas']

accuracy_train,accuracy_test=[],[]

for i in alphas:
    dtree=DecisionTreeClassifier(ccp_alpha=i)
    dtree.fit(X_train_small_pre, y_train_small)
    y_train_pred=dtree.predict(X_train_small_pre)
    y_test_pred=dtree.predict(X_test_small_pre)
    
    accuracy_train.append(accuracy_score(y_train_small,y_train_pred))
    accuracy_test.append(accuracy_score(y_test_small,y_test_pred))

sns.set()
plt.figure(figsize=(14,7))
sns.lineplot(y=accuracy_train,x=alphas,label="Train Accuracy")
sns.lineplot(y=accuracy_test,x=alphas,label="Test Accuracy")
plt.xticks(ticks=np.arange(0.00,0.02,0.01))
plt.show()
```

```python
## convert ideal_ccp_alpha from a series to float
ideal_ccp_alpha = alphas[accuracy_test.index(max(accuracy_test))]

#Create optimized Decision Tree

dt_classifier_opt = DecisionTreeClassifier(ccp_alpha=ideal_ccp_alpha)

pipe_dt_opt = Pipeline(steps=[('preprocessor', preprocessor),
                              ('classifier', dt_classifier_opt)])

# fit the Pipeline
dt_classifier_opt.fit(X_train_small_pre, y_train_small)

# Predict Labels
y_pred_dt_opt = dt_classifier_opt.predict(X_test_small_pre)

# Accuracy and Confusion matrix
print(f"Accuracy of the DT Model: {accuracy_score(y_test_small,y_pred_dt_opt)}")

plot_confusion_matrix(dt_classifier_opt, X_test_small_pre, y_test_small, display_labels= ["Not subscribed", "Subscribed"])  
plt.show()

#Accuracy of the DT Model: 0.776875
```

In this project, our objective is to find out which customers the bank should select to call based on customer profiles. We approached this business problem by following the computational analytics process with 5 main steps. We started with identification of the business problem, i.e., the success rate of a bank's marketing campaign was too low. In the problem analysis phase we delved deeper into the problem. We wanted to find out if there were any dependencies between the variables and problems and what kind of answer we would discover from the data to provide actionable insights for the bank to raise the success rate of the future marketing campaign. Specifically, we need to be able to classify which customers are likely to subscribe to a term deposit, using available customer data. This information would help the bank to target customers better and raise marketing campaign success rate, through customer preselection for contacting in their marketing campaigns. 

In design phase, we carried out analytics process planning, and we decided to use the bank client data’s attributes of 1-8 for minimal data requirement and to approach the solution by using several machine learning algorithms and comparing their performance.  4 selected machine learning algorithms were Decision Tree, Random Forest, Logistic Regression, and Support Vector Machine (SVM). 
In execution phase, we performed data extraction, transformation and analytics on these 4 machine learning models. The best prediction accuracy score on test data was Random Forest model with 67.87%, followed by Decision Tree with 66.75%. Logistic Regression’s accuracy score was 59.80%, while Support Vector Machine provided the least accuracy score at 57.44%.   We also did confusion matrix on the 4 models to find out how precisely each model can predict the target variable compared to true labels. In each of the used machine learning model, we also visualized top 15 features which impacted the most on the label. 

In random forest model, we discovered that age and balance were the 2 most important predictors for term deposit subscription. We did not know yet from random forest model what age group would be most likely to subscribe. In decision tree model, after pruning the tree, we got a much smaller tree, which showed us visually which customer segments have higher probabilities for  subscribing to a term deposit, depending on their age, education, and job. In logistic regression model, balance feature impacted the most on customer subscription likelihood, followed by age.  In support vector machine model, we further found out that retired, student and tertiary persons would likely subscribe the most to the term deposit, while blue collar and services persons would not. Balance and age factors in this model were not, however, as significant as they were in random forest and logistic regression models.

The answer we could provide the bank is that they should preselect people to contact based on balance, age, and job categories to improve the marketing campaign success rate. Further optimization of the models could provide deeper insights to this decision making. Especially the Decision Tree has high potential for customer segmentation.



In order to run more successful campaigns in the future the bank should prioritize customer segments that it includes in each campaign. Depending on resources and targeted number of new subscriptions the bank can start from most potential customer segments according to the machine learning models and move down to less potential groups until the goals are met or campaign time and resources are used. This segmentation and prioritization guarantee higher success rate than previous campaigns by excluding those customer segments that are not likely to subscribe and therefore saves resources for targeting more potential customer segments.

## Comparison of the Confusion Matrices

Performance and accuracy of the different algorithms is visualized in the form of confusion matrices. In a matrix we can see the true label of the answer (yes/no) on the vertical axis and the predicted label on the horizontal axis. When comparing each model’s confusion matrices we can first see that the models perform at quite similar rates, just like we could see in the Table 1 accuracy scores too. When observed more closely, we find out that the random forest had the most true positives out of all the models, meaning that it could most accurately predict which customers would subscribe to a term deposit. Decision tree was the most successful in predicting true negatives, meaning accurately predicting customers who would not subscribe when called. When it comes to other models and their true negatives, the results were quite similar to each other and all worse than the decision tree. Their true positives were not very accurate either and there was also more variation between the models. Especially the decision tree seems to be doing badly in finding accurate subscriptors. 

All in all, we can clearly see that our logistic regression and support vector machine models were the least successful in predicting the subscriptions, and it would not be wise to use those algorithms in the real world without some further development.
